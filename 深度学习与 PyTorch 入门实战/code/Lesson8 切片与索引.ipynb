## select by specific index 

```python
a.shape
```

```python
index = torch.tensor([0,2], dtype=torch.long)
```

```python
a.index_select(0,index).shape

# 选择 0 维度 的 0， 2 这两个
```

```python
idx = torch.tensor([0,1], dtype= torch.long)
a.index_select(1, idx).shape
```

```python
a.index_select(2, torch.arange(28)).shape
```

```python
a.index_select(2, torch.arange(8)).shape
```
